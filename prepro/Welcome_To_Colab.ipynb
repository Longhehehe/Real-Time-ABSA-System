```
import pandas as pd
import numpy as np
from google.colab import files
```

```
def processing(df):
  df['reviewContent'] = df['reviewContent'].replace('', np.nan)
  df = df.dropna(subset=['reviewContent'])
  return df
```

```
for i in range(11):
  input_path = f"concatenated_reviews_group_{i+1}.xlsx"
  df = pd.read_excel(input_path)
  df = processing(df)

  output_filename = f"processed_reviews_{i+1}.xlsx"
  df.to_excel(output_filename, index=False)

  files.download(output_filename)
```
